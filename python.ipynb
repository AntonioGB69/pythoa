```python
html_content = r"""
<!DOCTYPE html>
<html lang='pt-br'>
<head>
    <meta charset='UTF-8'>
    <meta name='viewport' content='width=device-width, initial-scale=1.0'>
    <title>Jogo de Adivinhação Maneiro</title>
    <style>
        body { font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif; display: flex; justify-content: center; align-items: center; height: 100vh; margin: 0; background-color: #2c3e50; color: white; }
        .container { background: #34495e; padding: 2rem; border-radius: 15px; box-shadow: 0 10px 25px rgba(0,0,0,0.3); text-align: center; width: 350px; }
        h1 { color: #1abc9c; margin-bottom: 0.5rem; }
        input { padding: 10px; border-radius: 5px; border: none; width: 60px; font-size: 1.2rem; text-align: center; }
        button { padding: 10px 20px; border-radius: 5px; border: none; background: #1abc9c; color: white; cursor: pointer; font-weight: bold; margin-top: 10px; }
        button:hover { background: #16a085; }
        #feedback { margin-top: 20px; font-weight: bold; min-height: 50px; }
        .hot { color: #e74c3c; }
        .warm { color: #f1c40f; }
        .cold { color: #3498db; }
        .stats { margin-top: 15px; font-size: 0.9rem; border-top: 1px solid #7f8c8d; padding-top: 10px; }
    </style>
</head>
<body>
    <div class='container'>
        <h1>🎮 Adivinhação</h1>
        <p>Tente um número entre 1 e 100</p>
        <input type='number' id='guessInput' min='1' max='100'>
        <br>
        <button onclick='checkGuess()'>Chutar!</button>
        <div id='feedback'></div>
        <div class='stats'>
            Tentativas: <span id='count'>0</span>/10<br>
            Recorde: <span id='best'>-</span>
        </div>
        <button id='resetBtn' style='display:none; background:#e67e22;' onclick='resetGame()'>Jogar Novamente</button>
    </div>

    <script>
        let secretNumber = Math.floor(Math.random() * 100) + 1;
        let attempts = 0;
        const maxAttempts = 10;
        let bestScore = localStorage.getItem('bestScore') || Infinity;

        if(bestScore !== Infinity) document.getElementById('best').innerText = bestScore;

        function checkGuess() {
            const input = document.getElementById('guessInput');
            const feedback = document.getElementById('feedback');
            const guess = parseInt(input.value);

            if (isNaN(guess) || guess < 1 || guess > 100) {
                feedback.innerText = 'Digite um número válido!';
                return;
            }

            attempts++;
            document.getElementById('count').innerText = attempts;

            const diff = Math.abs(guess - secretNumber);

            if (guess === secretNumber) {
                feedback.innerHTML = '<span style="color:#2ecc71">\uD83C\uDF89 ACERTOU!</span>';
                endGame(true);
            } else if (attempts >= maxAttempts) {
                feedback.innerHTML = `\u274C Game Over! Era ${secretNumber}`;
                endGame(false);
            } else {
                if (diff <= 5) feedback.className = 'hot', feedback.innerText = '\uD83D\uDD25 QUENTE!';
                else if (diff <= 15) feedback.className = 'warm', feedback.innerText = '\uD83C\uDF21\uFE0F Quentinho';
                else feedback.className = 'cold', feedback.innerText = '\u2744\uFE0F Frio!';
            }
            input.value = '';
            input.focus();
        }

        function endGame(won) {
            document.querySelector('button').disabled = true;
            document.getElementById('resetBtn').style.display = 'inline-block';
            if (won && attempts < bestScore) {
                bestScore = attempts;
                localStorage.setItem('bestScore', bestScore);
                document.getElementById('best').innerText = bestScore;
            }
        }

        function resetGame() {
            location.reload();
        }
    </script>
</body>
</html>
"""
with open('index.html', 'w', encoding='utf-8') as f:
    f.write(html_content)
print("Arquivo 'index.html' gerado com sucesso!")
```

```python
!pip install colorama
import random
import os
import json
from colorama import init, Fore, Style
import time

# Inicializa colorama para cores no terminal
init(autoreset=True)

# Arquivo para salvar recordes
RECORDE_FILE = 'recordes.json'

def carregar_recordes():
    if os.path.exists(RECORDE_FILE):
        with open(RECORDE_FILE, 'r') as f:
            return json.load(f)
    return {'melhor_tentativas': float('inf'), 'jogos_jogados': 0}

def salvar_recordes(recordes):
    with open(RECORDE_FILE, 'w') as f:
        json.dump(recordes, f)

def animacao_inicio():
    print(Fore.CYAN + Style.BRIGHT + """
    🎮 BEM-VINDO AO JOGO DE ADIVINHAÇÃO MANEIRO! 🎮
    """)
    time.sleep(0.5)
    print(Fore.YELLOW + "Eu pensei em um número de 1 a 100. Tente adivinhar!")
    print(Fore.GREEN + "Dicas: Quente (perto), Frio (longe). Boa sorte! 🔥❄️")
    time.sleep(1)

def dica(tentativa, numero):
    diff = abs(tentativa - numero)
    if diff == 0:
        return None
    elif diff <= 5:
        return Fore.RED + Style.BRIGHT + "🔥 QUENTE! Muito perto!"
    elif diff <= 15:
        return Fore.YELLOW + "🌡️ Quentinho!"
    else:
        return Fore.BLUE + "❄️ Frio! Tente outro número."

def animacao_vitoria(tentativas):
    print(Fore.GREEN + Style.BRIGHT + """
    🎉 PARABÉNS! Você acertou! 🎉
           .-1""""""""-.
         .'          '.
        /   ʕ•ᴥ•ʔ    \\
       : ,          : '
        `._         _.'
           `"'\"\"\"\"\"'`"
    """)
    print(f"Você precisou de {tentativas} tentativas! Incrível! 🏆")

def jogo():
    recordes = carregar_recordes()
    recordes['jogos_jogados'] += 1

    numero = random.randint(1, 100)
    tentativas = 0
    max_tent = 10  # Limite para desafio extra

    animacao_inicio()

    while tentativas < max_tent:
        try:
            palpite = int(input(Fore.MAGENTA + Style.BRIGHT + f"\nTentativa {tentativas + 1}/{max_tent}: "))
            tentativas += 1

            if palpite == numero:
                animacao_vitoria(tentativas)
                if tentativas < recordes['melhor_tentativas']:
                    recordes['melhor_tentativas'] = tentativas
                    print(Fore.YELLOW + f"Novo recorde: {tentativas} tentativas! 🌟")
                salvar_recordes(recordes)
                return True

            dica_msg = dica(palpite, numero)
            if dica_msg:
                print(dica_msg)
            else:
                print(Fore.CYAN + "Número inválido, entre 1 e 100!")

        except ValueError:
            print(Fore.RED + "Digite um número válido!")

    print(Fore.RED + Style.BRIGHT + f"Game Over! O número era {numero}. 😅")
    salvar_recordes(recordes)
    return False

def mostrar_recordes():
    recordes = carregar_recordes()
    print(Fore.CYAN + Style.BRIGHT + f"""
🏆 RECORDES ATUAIS 🏆
- Jogos jogados: {recordes['jogos_jogados']}
- Melhor: {recordes['melhor_tentativas']} tentativas (se inf, ainda sem recorde)
    """)

# Loop principal do jogo
if __name__ == "__main__":
    while True:
        mostrar_recordes()
        if input(Fore.WHITE + "\nJogar? (s/n): ").lower() == 's':
            jogo()
        else:
            print("Valeu por jogar! 👋")
            break
```

```python
import random
import ipywidgets as widgets
from IPython.display import display, clear_output

class JogoAdivinhacaoGUI:
    def __init__(self):
        self.numero_secreto = random.randint(1, 100)
        self.tentativas = 0
        self.max_tentativas = 10

        # Componentes da Interface
        self.titulo = widgets.HTML("<h2>🎮 Jogo de Adivinhação 🎮</h2>")
        self.instrucoes = widgets.Label(value="Tente adivinhar o número entre 1 e 100:")

        self.slider = widgets.IntSlider(value=50, min=1, max=100, step=1, description='Seu Palpite:', style={'description_width': 'initial'})
        self.botao_chutar = widgets.Button(description="Chutar!", button_style='primary', icon='check')
        self.botao_reiniciar = widgets.Button(description="Jogar Novamente", button_style='success', icon='refresh', layout=widgets.Layout(display='none'))

        self.progresso = widgets.IntProgress(value=0, min=0, max=10, description='Tentativas:', bar_style='info', orientation='horizontal')
        self.output = widgets.Output()

        # Eventos
        self.botao_chutar.on_click(self.processar_palpite)
        self.botao_reiniciar.on_click(self.reiniciar_jogo)

        # Layout
        self.layout_principal = widgets.VBox([
            self.titulo,
            self.instrucoes,
            self.slider,
            self.botao_chutar,
            self.progresso,
            self.botao_reiniciar,
            self.output
        ], layout=widgets.Layout(align_items='center', border='2px solid #444', padding='20px', width='400px'))

    def processar_palpite(self, b):
        self.tentativas += 1
        palpite = self.slider.value
        self.progresso.value = self.tentativas

        with self.output:
            clear_output()
            diff = abs(palpite - self.numero_secreto)

            if palpite == self.numero_secreto:
                print(f"🎉 PARABÉNS! Você acertou em {self.tentativas} tentativas!")
                self.finalizar_jogo(True)
            elif self.tentativas >= self.max_tentativas:
                print(f"❌ Game Over! O número era {self.numero_secreto}.")
                self.finalizar_jogo(False)
            else:
                if diff <= 5:
                    print("🔥 QUENTE! Você está muito perto!")
                elif diff <= 15:
                    print("🌡️ Quentinho! Está chegando lá.")
                else:
                    print("❄️ Frio! Tente outro número.")

                if palpite < self.numero_secreto:
                    print("💡 Dica: O número é MAIOR.")
                else:
                    print("💡 Dica: O número é MENOR.")

    def finalizar_jogo(self, venceu):
        self.botao_chutar.disabled = True
        self.slider.disabled = True
        self.botao_reiniciar.layout.display = 'block'
        self.progresso.bar_style = 'success' if venceu else 'danger'

    def reiniciar_jogo(self, b):
        self.numero_secreto = random.randint(1, 100)
        self.tentativas = 0
        self.progresso.value = 0
        self.progresso.bar_style = 'info'
        self.botao_chutar.disabled = False
        self.slider.disabled = False
        self.botao_reiniciar.layout.display = 'none'
        with self.output:
            clear_output()
            print("Novo jogo iniciado! Boa sorte.")

    def exibir(self):
        display(self.layout_principal)

# Iniciar o jogo
jogo_gui = JogoAdivinhacaoGUI()
jogo_gui.exibir()
```
